```python
# imports
import deepxde as dde
import numpy as np
import scipy as sp
```

```python
# Defining PDE to model a physical system. In this case, we model 1D advection through time.
## Advection: the transport of a substance by bulk motion of a fluid. 
### EX: Pollutants carried downstream by the bulk water flow of a river
## The PDE models the concentration of a substance carried by a fluid as a function of the distance from the source of the carried substance (x)


# Formulation of the residuals of the partial differential equation for advection diffusion equation
def pde(x, y):
    # x is the vector of inputs [x_downstream, t] where x_downstream is the distance from source and t is time
    ## j = 0 means the partial with respect to x_downstream
    ## j = 1 means the partial with respect to t 
    # y is the output, concentration 
    dy_t = dde.grad.jacobian(y, x, j=1)
    dy_x = dde.grad.jacobian(y, x, j=0)
    dy_xx = dde.grad.hessian(y, x, j=0)
    return (dy_t - D*dy_xx + u*dy_x)

# Definition of the initial condition
def initial(x):
    return 0

# Analytical solution (Ogata Banks equation)
## only used to provide a benchmark to test model performance, not used to train the model 
## if such an analytical solution is not available, one may have to compare the output with the output of a numerical solution (e.g., finite elements)
## a lower loss function is not necessarily associated with a lower error between the model output and the true system behavior. 
## And there are issues with overfitting that you would want to address too by holding out a test set of data points on which you compute accuracy.

def analytical(X_and_T):
    x_saved = X_and_T
    x_=np.array(X_and_T[:, 0:1])
    t = np.array(X_and_T[:, 1:])
    c_0=.5*sp.special.erfc(((x_+0.5)-u*(t+0.5))/(2*(D*(t+0.5))**.5))+1/2*np.exp(u*(x_+0.5)/D)*sp.special.erfc(((x_+0.5)+u*(t+0.5))/(2*(D*(t+0.5))**.5))
    c_0_high_peclet=.5*sp.special.erfc(((x_+0.5)-u*(t+0.5))/(2*(D*(t+0.5))**.5))
    c_0=np.nan_to_num(c_0,nan=c_0_high_peclet)
    return c_0*c_feed

# Functions to locate the boundaries
def boundary_l(x, on_boundary):
    return on_boundary and np.isclose(x[0], -0.5)
def boundary_r(x, on_boundary):
    return on_boundary and np.isclose(x[0], 0.5)

# Definition of the variables
u_real=0.30686887
D_real=0.072
L_real=86
T_real=200

# Peclet number
Pe = L_real * u_real / D_real
print('Peclet number =', Pe)

# Normalization (space and time are normalized to [-0.5, 0.5] for the )
## he mentioned in the lecture that if you change the activation function, you would want to check how the backend actually implements 
## the activation function check the domain of the function
u=u_real*T_real/L_real
D=D_real*T_real/L_real**2
c_feed=1 # fixed upstream source concentration 

# Definition of the domain
# spatial domain
geom = dde.geometry.Interval(-0.5, 0.5)
# temporal domain
timedomain = dde.geometry.TimeDomain(-0.5, 0.5)
# spatio-temporal domain
geomtime = dde.geometry.GeometryXTime(geom, timedomain)

# Definition of the boundary and initial conditions
## you can use other types of boundary conditions too, check the DeepXDE examples for PINN forward
bc_l = dde.icbc.DirichletBC(geomtime, lambda X: c_feed, boundary_l)
bc_r = dde.icbc.DirichletBC(geomtime, lambda x: 0, boundary_r)
ic = dde.icbc.IC(geomtime, initial, lambda _, on_initial: on_initial)

# Data training points (if we use them):
observe_x = np.vstack((np.linspace(-0.5,0.5,100), np.ones(100)*0.5)).T
observe_y = dde.icbc.PointSetBC(observe_x, analytical(observe_x), component=0)

# Define the partial differential equation in the format expected by DeepXDE
data = dde.data.TimePDE(
    geomtime, #Domain geometry
    pde, #Type of problem
    [bc_l, bc_r, ic, observe_y], #Residuals to minimize. Data points here are not being used because the loss weight in model.compile is 0.
    num_domain=1000, #Number of residual points in the spacetime domain, sampled uniformly. Increasing this gives more training points.
    num_boundary=100, #Number of residual points in the boundary condition
    num_initial=100, #Number of residual points in the initial condition
    solution=analytical, #Criterion used for the test metric 
    num_test=20, #Number of test points
)
```

```python
# Properties of the neural network
## Assumes a fully-connected neural network
## input layer is size 2 b/c we have x and t
## output layer is size 1 b/c we output the concentration given input (x, t)
layer_size = [2] + [32] * 4 + [32] + [1] #Size of the neural network. This forms a list with the number of neurons in each layer. First and last number must coincide with the number of input and output parameters respectively
activation = "sigmoid" # They are the activation functions from tensorflow as appearing in https://www.tensorflow.org/api_docs/python/tf/keras/activations
initializer = "Glorot normal" # They are the initializers from deepxde for tensorflow as can be seen in "def initializer_dict_tf():" in https://deepxde.readthedocs.io/en/latest/_modules/deepxde/nn/initializers.html
net = dde.nn.FNN(layer_size, activation, initializer)
```

```python
#Training the model:
model = dde.Model(data, net)
model.compile(
    "adam", #Optimizer, some other options include "L-BFGS-B". 
    ## You could also change the optimizer during training as mentioned by Prof. Tartakovsky is done in practice for many problems 
    ## (b/c SGD methods tend not to converge, but the loss oscillates), so often they switch to a quasi-Newtonian optimizer at the end of training.
    lr=0.001, #Learning rate. 0.001 is the default for Adam. Others optimizers like L-BFGS-B don't require this parameter
    metrics=["l2 relative error"], #Test metric
    loss_weights=[10,1,1,1,0] # Weights of each term in the loss function. 
    ## Loss weights are weights associated with: [PDE (residual), bc_l, bc_r, ic, observe_y] since that is the order they are given when we define the TimePDE datastructure
    ## Since the last term is 0 here, we do not use given measurements (observe_y) to compute any loss to train the model.
    )

losshistory, train_state = model.train(epochs=20000)
dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```

```python
# plot the PINN output vs the analytical solution
import numpy as np
import matplotlib.pyplot as plt
T_plot=0.5
N_x=100

Deltax=1/N_x
xs=np.arange(-N_x/2,N_x/2)*Deltax
X_and_T=np.array([[x,T_plot] for x in xs])
X=np.array(X_and_T[:, 0:1])

fig, ax = plt.subplots()
ax.plot(X,analytical(X_and_T),color='blue',label='Analytical')
ax.plot(X,model.predict(X_and_T),color='red',label='PINN')
ax.set_xlabel('Normalized x')
ax.set_ylabel('Concentration')
plt.grid()
plt.legend()
```

```python
# computer L2 error between PINN outputs and analytical solution across all time points
N_t=100
Deltat=1/N_t
ts=np.arange(-N_t/2+1,N_t/2+1)*Deltat
X_and_T=np.array(np.meshgrid(xs,ts)).reshape(N_t*N_x,2)

rel_L2_error=np.sum((analytical(X_and_T)-model.predict(X_and_T))**2)/np.sum((analytical(X_and_T))**2)
print('Relative L2 error =',rel_L2_error)
```
